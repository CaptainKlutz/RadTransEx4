```python
import numpy as np
import matplotlib.pyplot as plt
def findW(z,H,thetaS,sigma,n):
    W=np.exp(sigma*n*(1/np.cos(2*np.pi*thetaS/360))*(z/H)*np.exp(-z/H))
    return W
H=[8000,2000]
length=400
z=np.linspace(0,50000,num=length)
W=np.ones(length)
thetaS=30
sigma=1E-21
n=1E16
fig,ax=plt.subplots(figsize=(10,10))
for i in range (0,2):
    for x in range (0,length):
        W[x]=findW(z[x],H[i],thetaS,sigma,n)
    ax.plot(W,z,label="Scale height H="+str(H[i]))
ax.grid(b=None,which="major",axis="both")
ax.set_ylim([0,50000])
ax.set_ylabel("Height in Metres")
ax.set_xlabel("Absorption rate W")
ax.set_title("Absorption rate as a function of height")
ax.legend()
ax.ylim=(0,50000)
```
